```
#importing the necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```
#loading the dataset
iris = sns.load_dataset('iris')
iris.head()
```

```
Counter(iris['species'])
```

```
#univariate data plotting
#Histogram
sns.distplot(iris['sepal_length'])
plt.show()
```

```
plt.close()
```

```
plt.subplot(221)
sns.distplot(iris['sepal_length'])
plt.subplot(222)
sns.distplot(iris['petal_length'])
plt.subplot(223)
sns.distplot(iris['sepal_width'])
plt.subplot(224)
sns.distplot(iris['petal_width'])
plt.show()
```

```
plt.close()
```

```
#Bivariate Analysis
#Scatter plot
sns.jointplot(data = iris, x = 'sepal_length', y = 'petal_length')
plt.show()
```

```
sns.jointplot(x = iris['petal_length'], y = iris['petal_width'], kind = 'hex', color = 'green')
plt.show()
```

```
#Pair Plot
sns.pairplot(iris)
plt.show()
```

```
#Strip Plot
sns.stripplot(x = 'species', y = 'sepal_length', data = iris)
plt.show()
```

```
#Swarm Plot
sns.swarmplot(x = 'species', y = 'sepal_length', data = iris)
plt.show()
```
